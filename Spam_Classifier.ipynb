```python
# Load the mbox file
spam_csv = 'spam.csv'

# Open CSV file to write
with open(spam_csv, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Body'])

    for message in spam:
        try:
            # Get the email body (plain text only)
            body = ''
            if message.is_multipart():
                for part in message.walk():
                    if part.get_content_type() == 'text/plain':
                        body += part.get_payload(decode=True).decode(errors='ignore')
            else:
                body = message.get_payload(decode=True).decode(errors='ignore')

            writer.writerow([body.strip()])
        except Exception as e:
            print(f"Skipped one message due to error: {e}")
```

```python
# to lowercase
emails['text']=emails['text'].str.lower()
```

```python
# removing the html tags
for i in range(len(emails)):
    emails['text']=emails['text'].str.replace('<.*?>',' ')
```

```python
#remove \n
for i in range(len(emails)):
    emails['text']=emails['text'].str.replace('\n',' ')
```

```python
#remove \r
for i in range(len(emails)):
    emails['text']=emails['text'].str.replace('\r',' ')
```

```python
# remove urls
for i in range(len(emails)):
    emails['text']=emails['text'].str.replace('http\S+|www.\S+',' ')
```

```python
# remove punctuation
exclude=string.punctuation
def removePun(text):
    for char in exclude:
        text=text.replace(char,' ')
    return text
```

```python
for i in range(len(emails)):
    emails.at[emails.index[i], 'text'] = removePun(emails['text'].iloc[i])
```

```python
emails['text'] = emails['text'].str.lower().str.replace(r'[^\w\s]', ' ', regex=True)
```

```python


spell = SpellChecker()

def correct_spell_pyspellchecker(text):
    words = text.split()
    corrected_words = [
        spell.correction(word) if spell.correction(word) is not None else word
        for word in words
    ]
    return ' '.join(corrected_words)
```
